```julia
using CairoMakie, LinearAlgebra, NBInclude, LaTeXStrings
```

```julia
@nbinclude("set_makie_defaults.ipynb")
```

```julia
function f(u)
    x, y = u
    return Point2f(-3*x + 4*y, -2*x + y)
end

fig = Figure(size = 0.8 .* (800, 600))
ax = Axis(fig[1, 1], xlabel = L"y_1", ylabel = L"y_2", xlabelsize = 18, ylabelsize = 18, xticklabelsize = 0, yticklabelsize = 0)

streamplot!(ax, f, -4.0..4.0, -4.0..4.0, colormap = :plasma)

#equilibrium point
#scatter!(ax, [0.0], [0.0], color = :red, markersize = 15)

# Display the plot
display(fig)
save("phase_portrait.png", fig) # Save as SVG for high quality
```
