```julia
function simpson(f; interval::Tuple, N::Int)
    """
    Implementation of Composite Simpson's 1/3 quadrature rule for approximating definite integrals: 

    ∫_{a}^{b} f(x) dx ≈ Δx/3 * [f(x₁) + f(x_{2N+1}) + 4*S_even + 2*S_odd],     Δx = (b-a)/(2N) 

    where: 
    N  = # of "parabolic panels" [x_j, x_{j+2}]  
    S_even = f(x₂) + f(x₄) + f(x₆) + ... + f(x_{2N})
    S_odd  = f(x₃) + f(x₅) + f(x₇) + ... + f(x_{2N-1})

    NOTE: In some quadrature libraries, 'N' refers to the number of sub-intervals [x_j, x_{j+1}]. In this 
    function, N is the number of times that basic Simpson's 1/3 rule is applied. 

    PARAMETERS
    ----------
    f :: the function to be integrated; can be either scalar-valued or vector-valued. 
    interval :: the interval on which to integrate; must be finite. 
    N :: the number of parabolic panels [x_j, x_{j+2}], i.e., the # of times the basic Simpson's 1/3 rule is applied. 

    RETURNS
    -------
    I :: the Simpson's 1/3 rule approximation of ∫_{a}^{b} f(x) dx (either a scalar or a vector, depending on f)
    """

    a, b = interval
    a < b || throw(ArgumentError("Invalid interval: left endpoint must be less than the right endpoint."))
    N ≥ 1 || throw(ArgumentError("N must be a positive integer."))
  
    #Mesh spacing
    Δx = (b-a) / (2*N) 

    #Nodes (x[j] = x_j = a + (j-1)Δx, j = 1,2,...,2N+1) 
    x = range(a, step = Δx, length = 2*N + 1)
    y = f.(x)
    
    S_even = sum(y[2:2:2*N]) 
    S_odd  = N ≥ 2 ? sum(y[3:2:2*N-1]) : zero(y[1]) 
        
    return (Δx/3) * (y[1] + 4*S_even + 2*S_odd + y[end])
end 
```
