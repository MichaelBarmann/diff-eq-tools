```julia
using CairoMakie, LaTeXStrings
```

```julia
function plot_sol_1D(u_num::Vector{T}, t::Vector{T}; u_ex::Union{Vector{T},Nothing} = nothing, size = (400,400), 
                     title = "", markersize = 4, linewidth = 4, legend_pos = :rt) where T 

    fig = Figure(size = size)

    #Create plot 
    ax = Axis(fig[1,1], xlabel = L"t", ylabel = L"u", limits = (t[1], t[end], minimum(u_num), maximum(u_num)), title = title)

    #Plot exact solution (if given) 
    if u_ex != nothing
        lines!(ax, t, u_ex, linewidth = linewidth, color = :blue, label = "exact")
    end 

    #Plot numerical solution 
    scatter!(ax, t, u_num, markersize = markersize, color = :orange, label = "numerical")

    #Add legend 
    axislegend(ax, position = legend_pos)
    
    display(fig)
end 
```

```julia
function plot_sol_2D(U::Vector{Vector{T}}, t::Vector{Float64}; U_exact::Union{Vector{Vector{T}},Nothing} = nothing, size = (400,400), 
                     title = "", markersize = 4, linewidth = 4, legend_pos = :rt) where T 

    """
    Plot the solution to a two-component ODE system U' = f(U,t,p), U(0) = U₀. 
    The solution should be in the form U = [U_0, U_1,...,U_M], where U_i is the solution at the ith time step
    with U_i being a vector of length 2. Likewise for the exact solution. 
    """

    fig = Figure(size = size)

    u_num = getindex.(U, 1)
    v_num = getindex.(U, 2)
    u_ex = U_exact != nothing ? getindex.(U_exact, 1) : nothing
    v_ex = U_exact != nothing ? getindex.(U_exact, 2) : nothing 

    u_min = min(minimum(u_num), minimum(v_num))
    u_max = max(maximum(u_num), maximum(v_num))
    ymin = u_min - 0.10 * abs(u_min)
    ymax = u_max + 0.10 * abs(u_max)
     
    #Create plot 
    ax = Axis(fig[1,1], xlabel = L"t", limits = (t[1], t[end], ymin, ymax), title = title)

    #Plot exact solution (if given) 
    if U_exact != nothing
        lines!(ax, t, u_ex, linewidth = linewidth, color = :cyan, label = L"u_{\textrm{exact}}")
        lines!(ax, t, v_ex, linewidth = linewidth, color = :orange, label = L"v_{\textrm{exact}}")
    end 

    #Plot numerical solution 
    scatter!(ax, t, u_num, markersize = markersize, color = :purple, label = L"u_{\textrm{numerical}}")
    scatter!(ax, t, v_num, markersize = markersize, color = :red, label = L"u_{\textrm{numerical}}")
    
    #Add legend 
    axislegend(ax, position = legend_pos)
    
    display(fig)
end 
```
