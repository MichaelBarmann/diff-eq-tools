```julia
function RK45_step(f, u_i, t_i, h, p, reltol = 1e-3, abstol = 1e-6)

    """
    Given the numerical solution z_i at some time t_i, compute the solution u_{i+1} at t_{i+1} = t_i + h
    where h is chosen so that the condition norm(u_new - w_new) < reltol * norm(u_new) + abstol is satisfied. 
    """

    e_i = Inf 
    tol_combined = 0.0
       
    while e_i ≥ tol_combined 

        s1 = f(u_i, p, t_i)
        s2 = f(u_i + 0.25 * h * s1, p, t_i + 0.25*h)
        s3 = f(u_i + (3/32) * h * s1 + (9/32) * h * s2, p, t_i + (3/8) * h )
        s4 = f(u_i + (1932/2197) * h * s1 - (7200/2197) * h * s2 + (7296)/(2197) * h * s3, p, t_i + (12/13) * h)
        s5 = f(u_i + (439/216) * h * s1 - 8*h*s2 + (3680/513)*h*s3 - (845/4104)*h*s4, p, t_i + h)
        s6 = f(u_i - (8/27)*h*s1 + 2*h*s2 - (3544/2565)*h*s3 + (1859/4104)*h*s4 - (11/40)*h*s5, p, t_i + 0.5*h)

        #Order 4 approximation of u(t_i + h)
        w_new = u_i + h * ((25/216)*s1 + (1408/2565)*s3 + (2197/4104)*s4 - (1/5)*s5)

        #Order 5 approximation of u(t_i + h) 
        u_new = u_i + h * ((16/135)*s1 + (6656/12825)*s3 + (28561/56430)*s4 - (9/50)*s5 + (2/55)*s6)

        e_i = norm(u_new - w_new)
        
        #Update tol_combined
        tol_combined = reltol * norm(u_new) + abstol 
        
        if e_i < tol_combined
            return u_new, t_i + h, h 
        else
            #h_new = 0.8 * h * (e_i / norm(u_i))^(0.2)
            h *= 0.8 * (tol_combined / e_i)^0.2
        end

    end
end
```
