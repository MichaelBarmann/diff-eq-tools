```julia
function midpoint(f; interval::Tuple, N::Int) 
    """
    The Midpoint Rule for approximating a definite integral: 
    ∫_[a,b] f(x) dx ≈ \frac{b-a}{N} ∑_{j=1}^{N} f(x_j) where x_j = a + (j-1/2)(b-a)/M, j = 1,2,...,M.

    PARAMETERS
    ----------
    f :: a scalar-valued or vector-valued function defined on an interval. 
    interval :: the interval of integration (must be finite) 
    N :: the number of quadrature nodes 

    RETURNS
    -------
    I :: (scalar or vector) The approximate integral \frac{1}{N} ∑_{j=1}^{N} f(x_j). 
    """

    a, b = interval

    #Input validation 
    a < b || throw(ArgumentError("Invalid interval. Left endpoint must be less than right endpoint."))
    N ≥ 1 || throw(ArgumentError("N must be a positive integer."))

    #Node spacing
    Δx = (b-a) / N 

    #Nodes: x[j] = xⱼ = a + Δx(j-1/2), j = 1,...,N 
    x = range(a + Δx/2, step = Δx, length = N)

    return Δx * sum(f.(x))
end 
```
