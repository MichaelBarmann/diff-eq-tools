```julia
using Test, NBInclude
```

```julia
@nbinclude("trapezoidal.ipynb")
```

```julia
#Test `trapezoidal()` 
@testset "Composite Trapezoidal Rule" begin
    f1(x) = x             #Exact integral: 1/2
    @test trapezoidal(f1; interval = (0,1), N = 1) ≈ 1/2 

    f2(x) = 5x + 2        #Exact integral: 9/2 
    @test trapezoidal(f2; interval = (0,1), N = 1) ≈ 9/2

    f3(x) = x^2 - 2x + 5  #Exact integral: 13/3
    @test trapezoidal(f3; interval = (0,1), N = 50) ≈ 13/3    atol = 1e-3

    f4(x) = sin(x)        #Exact integral: 2  
    @test trapezoidal(f4; interval = (0,π), N = 50) ≈ 2       atol = 1e-3

    f5(x) = exp(x)        #Exact integral
    @test trapezoidal(f5; interval = (0,log(3)), N = 50) ≈ 2  atol = 1e-3

    f6(x) = [x, 2x + 1]
    @test trapezoidal(f6; interval = (2,3), N = 1) ≈ [5/2, 6]   

    f7(x) = [x, sin(x)]
    @test trapezoidal(f7; interval = (0,π), N = 50) ≈ [(π^2)/2, 2]  atol = 1e-3
end 
```
