```julia
function boole(f; interval::Tuple, N::Int)
    """
    Boole's Rule for approximating definite integrals: 
    ∫_{a}^{b} f(x) dx = (b-a)/(90N) * [7*(f(x₁) + f(x_{4N+1})) + 32*S₁ + 12*S₂ + 14*S₃]
    
    PARAMETERS
    ----------
    f :: integrand 
    interval :: interval on which to integrate f
    N :: number of times the basic Boole's Rule is to be applied

    RETURNS
    -------
    integral :: numerical approximation to ∫_{a}^{b} f(x) dx (error is O(h^6)). 
    """

    #Input validation
    a, b = interval
    a < b || throw(ArgumentError("Invalid interval: left endpoint must be less than right endpoint."))
    N ≥ 1 || throw(ArgumentError("N must be a positive integer."))

    #node spacing
    Δx = (b-a) / (4*N) 

    #nodes 
    x = range(a, step = Δx, length = 4*N + 1)

    #function values 
    y = f.(x)
    
    #sums 
    S₁ = sum(y[2:2:4*N])
    S₂ = sum(y[3:4:4*N-1])
    S₃ = N ≥ 2 ? sum(y[5:4:4*N-3]) : zero(y[1])

    integral = (2/45 * Δx) * (7*(y[1] + y[end]) + 32*S₁ + 12*S₂ + 14*S₃)
    return integral
end 
```
