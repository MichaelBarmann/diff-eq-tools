### trapezoidal.ipynb

```julia
function trapezoidal(f; interval::Tuple, N::Int)
    """
    The composite Trapezoidal Rule for approximating a definite integral. 
        ∫_{a}^{b} f(x) dx ≈ (b-a)/N *([f(a) + f(b)]/2 + ∑_{j=1}^{N-1} f(x_j)) 
    """
    a, b = interval
    a < b || throw(ArgumentError("Invalid interval: must have [left endpoint] < [right endpoint]."))

    N > 0 || throw(ArgumentError("N must be a positive integer."))
    h = (b-a)/N

    #N + 1 evenly spaced quadrature nodes 
    x = range(a, b; length = N+1) 
    y = f.(x)

    interior_sum = N ≥ 2 ? sum(y[2:end-1]) : zero(y[1])
    return h * ((y[1] + y[end]) / 2 + interior_sum)
end 
```
