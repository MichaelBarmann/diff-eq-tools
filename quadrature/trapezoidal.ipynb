### trapezoidal.ipynb

```julia
function trapezoidal(f; interval::Tuple, N::Int)
    """
    The composite Trapezoidal Rule: 
        ∫_{a}^{b} f(x) dx ≈ (b-a)/N *([f(a) + f(b)]/2 + ∑_{j=1}^{N-1} f(x_j)) 
    """
    a, b = interval
    h = (b-a)/N

    #N + 1 evenly spaced quadrature nodes 
    x = collect(range(a, step = h, length = N + 1))
    y = f.(x)

    return h*((y[1] + y[end])/2 + sum(y[2:end-1])) 
end 
```
